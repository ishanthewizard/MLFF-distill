```python
import os
import shutil
import re

# Set the main directory path
main_directory = "/home/sanjeevr/MLFF-distill/data/labels_unlocked/EScAIP_labels/iodine/force_jacobians"  # Change this to your actual directory path

# Regular expression to match the files like "data1234.lmdb" and "lock1234.lmdb"
data_pattern = re.compile(r"data(\d{4})\.lmdb")
lock_pattern = re.compile(r"lock(\d{4})\.lmdb")

# List all files in the main directory
files = os.listdir(main_directory)

# Identify unique 4-digit numbers
unique_numbers = set()

for file in files:
    data_match = data_pattern.fullmatch(file)
    lock_match = lock_pattern.fullmatch(file)

    if data_match:
        unique_numbers.add(data_match.group(1))  # Extract the 4-digit number
    elif lock_match:
        unique_numbers.add(lock_match.group(1))  # Extract the 4-digit number

# Process each unique number to restore the original structure
for number in unique_numbers:
    folder_name = f"data.{number}.lmdb"
    folder_path = os.path.join(main_directory, folder_name)

    # Create the folder if it doesn't exist
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Created folder: {folder_path}")

    # Define file paths
    data_file = os.path.join(main_directory, f"data{number}.lmdb")
    lock_file = os.path.join(main_directory, f"lock{number}.lmdb")

    new_data_file = os.path.join(folder_path, "data.lmdb")
    new_lock_file = os.path.join(folder_path, "lock.lmdb")

    # Move files back into their respective folders
    if os.path.exists(data_file):
        shutil.move(data_file, new_data_file)
        print(f"Moved: {data_file} -> {new_data_file}")

    if os.path.exists(lock_file):
        shutil.move(lock_file, new_lock_file)
        print(f"Moved: {lock_file} -> {new_lock_file}")

print("Reorganization complete.")
```
